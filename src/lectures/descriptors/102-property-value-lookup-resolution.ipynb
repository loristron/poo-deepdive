# Property Lookup resolution

```python
class IntegerValue:

    def __set__(self, instane, value):
        print('__set__ called...')

    def __get__(self, instance, owner_class):
        print('__get__ called')
```

```python
class Point:
    x = IntegerValue()
```

```python
p = Point()
```

```python
p.x = 100
```

```python
p.x
```

```python
p.__dict__
```

```python
p.__dict__['x'] = 'hello'
p.__dict__
```

```python
p.x
```

Python ignore what's in the isntance dictionary and goes directly into the descriptior. This behavior is only True when we're dealing with __data descriptors__

## Example of non-data descriptor

```python
class TimeUTC:
    def __get__(self, instance, owner_class):
        print('__get__ called...')


class Logger:
    current_time = TimeUTC()
```

```python
l = Logger()
l.current_time
```

```python
l.__dict__
```

```python
l.__dict__['current_time'] = 'hello'
```

```python
l.current_time
```

For non-data descriptors the instance dictionary shadows the descriptors attributes

```python
del l.__dict__['current_time']
```

```python
l.current_time
```

When the key is not present in the dictionray, uses the one in the get method

```python
class ValidString:
    def __init__(self, min_len=None):
        self.min_len = min_len

    def __set_name__(self, owner_class, property_name):
        self.property_name = property_name
        
    def __set__(self, instance, value):
        if not isinstance(value, str):
            raise ValueError(f'{self.property_name} must be a string')
        
        if self.min_len is not None and len(value) < self.min_len:
            raise ValueError(f'{self.property_name} field must be at least {self.min_len} characters long')
        
        instance.__dict__[self.property_name] = value # directly propagating the value in the instance dictionary

    def __get__(self, instance, owner_class):
        if instance is None:
            return self 
        
        print(f'calling __get__ for {self.property_name}')
        return instance.__dict__.get(self.property_name, 'Property not setted')
        
```

```python
class Person:
    first_name = ValidString(1)
    last_name = ValidString(2)
```

```python
p = Person()
p.__dict__
```

```python
p.first_name = 'Alex'
```

```python
p.__dict__
```

```python
p.first_name
```

```python
class ValidString:
    def __init__(self, min_len=None):
        self.min_len = min_len

    def __set_name__(self, owner_class, property_name):
        self.property_name = property_name
        
    def __set__(self, instance, value):
        print(f'__set__ method called for {self.property_name}')
        if not isinstance(value, str):
            raise ValueError(f'{self.property_name} must be a string')
        
        if self.min_len is not None and len(value) < self.min_len:
            raise ValueError(f'{self.property_name} field must be at least {self.min_len} characters long')
        
        # instance.__dict__[self.property_name] = value # directly propagating the value in the instance dictionary
        setattr(instance, self.property_name, value)

    def __get__(self, instance, owner_class):
        if instance is None:
            return self 
        
        print(f'calling __get__ for {self.property_name}')
        # return instance.__dict__.get(self.property_name, 'Property not setted')
        return getattr(instance, self.property_name, 'None') # equivalent on trying to return self.first_name 
        
```

```python
class Person():
    name = ValidString(1)
```

```python
p = Person()
p.first_name = 'Alex'
```

```python
p.first_name
```

```python
p.name = 'Julia'
```
